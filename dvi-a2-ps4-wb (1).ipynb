### Question 4
<h4>Visualize the matches played by each team and find out which team played the maximum number of matches in IPL. <br><br>
    Write the python code in the below cell to create appropriate visual to perform the above task.
<h4>Answer in markdown cells below the visual <br><br>
    1.Summarize your findings from the visual.<br>
    2.The reason for selecting the chart type you did. <br>
    3.Mention the pre-attentive attributes used.(atleast 2)<br>
    4.Mention the gestalt principles used.(atleast 2)<br>

```python
# Concatenate team1 and team2 columns, and then count the number of matches played by each team
matches_played = pd.concat([df2['team1'], df2['team2']]).value_counts()

# Find the team that played the maximum number of matches
max_matches_team = matches_played.idxmax()
max_matches = matches_played[max_matches_team]

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=matches_played.values, y=matches_played.index, palette='Set2')
plt.xlabel('Number of Matches Played')
plt.ylabel('Team')
plt.title('Matches Played by Each IPL Team')
plt.annotate(f"{max_matches_team} ({max_matches} matches)", xy=(max_matches, matches_played.index.get_loc(max_matches_team)), xytext=(max_matches + 20, matches_played.index.get_loc(max_matches_team)), arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5'), fontsize=12, color='red')
plt.xticks(rotation=0)
plt.show()
```

```python
# Findings:
# 1. Mumbai Indians (MI) played the maximum number of matches in the IPL.
# 2. The chart type selected is a horizontal bar plot because it effectively displays the count of matches played by each team, making it easy to compare.
# 3. Pre-attentive attributes used:
#    - Position: The position of bars on the y-axis represents the teams, while the length of bars on the x-axis represents the number of matches played.
#    - Color: A distinct color palette ('Set2') is used to differentiate between teams.
# 4. Gestalt principles used:
#    - Proximity: Teams with similar colors and names are grouped together, allowing viewers to easily associate them.
#    - Similarity: Similar colors are used for related elements (teams) to make them visually coherent.

# In summary, the visual shows the number of matches played by each IPL team, with Mumbai Indians (MI) having the highest count of matches played over the years.
```

```python
# Count the number of matches played at each venue
venue_counts = df2['venue'].value_counts()

# Find the venue with the most number of matches
most_matches_venue = venue_counts.idxmax()
most_matches = venue_counts[most_matches_venue]

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=venue_counts.values, y=venue_counts.index, palette='Set3')
plt.xlabel('Number of Matches Played')
plt.ylabel('Venue')
plt.title('Number of Matches Played at Different IPL Venues')
plt.annotate(f"{most_matches_venue} ({most_matches} matches)", xy=(most_matches, venue_counts.index.get_loc(most_matches_venue)), xytext=(most_matches + 20, venue_counts.index.get_loc(most_matches_venue)), arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5'), fontsize=12, color='red')
plt.xticks(rotation=0)
plt.show()

```

```python
# Findings:
# 1. Eden Gardens had the most number of matches played among all IPL venues.
# 2. The chart type selected is a horizontal bar plot because it effectively displays the count of matches played at each venue, making it easy to compare.
# 3. Pre-attentive attributes used:
#    - Position: The position of bars on the y-axis represents the venues, while the length of bars on the x-axis represents the number of matches played.
#    - Color: A distinct color palette ('Set3') is used to differentiate between venues.
# 4. Gestalt principles used:
#    - Proximity: Venues with similar colors and names are grouped together, allowing viewers to easily associate them.
#    - Similarity: Similar colors are used for related elements (venues) to make them visually coherent.
```
